# VICTORY-DRAW-LOSS
In order to identify if possession heads to better results, it is necessary to know what was the result of each match for each team. With wdl function, we check the goals column to determine if the team of every row won, lost, or draw.

```python
def wdl(x):
    if (x+2) % 2 == 0:                                  #Distinction between Home and Visitor depending on its index position
        if ligas['Goals'][x] > ligas['Goals'][x+1]:     #(Home = odd, Visitor = even)
            x = 'Victory'                               #More goals than the immediate next row --> Victory
            return x                                    #Less goals than the immediate next row --> Loss
        elif ligas['Goals'][x] < ligas['Goals'][x+1]:   #Same goals --> Draw
            x = 'Loss'
            return x
        else:
            x = 'Draw'
            return x
    if (x+2) % 2 != 0:                                  #Same but with the previous row
        if ligas['Goals'][x] > ligas['Goals'][x-1]:
            x = 'Victory'
            return x
        elif ligas['Goals'][x] < ligas['Goals'][x-1]:
            x = 'Loss'
            return x
        else:
            x = 'Draw'
            return x

ligas['resultados'] = range(0, len(ligas))              #Create row with the same values as the index to iterate
ligas['resultados'] = ligas['resultados'].apply(wdl)    #Method.apply() to use it in every value of the column
ligas.head(4)
```

## Duplicate teams: same league

When teams with the same name play in the same league, we will check the match name to make them different again. To get that, it was necessary to create a column: 'Same name and league', and a function 'snal'.

We have two cases of that, Monaco-Montpellier (both MON), and Real Valladolid-Rayo Vallecano (both RV)

```python
ligas['Same name and league'] = range(0, len(ligas))        #We needed to iterate over a column with the same values as the index

def snal(x):
    if ligas['Teams'][x] != 'MON' and ligas['Teams'][x] != 'RV':                #Monaco keeps MON so we will call Montpellier MNT
        return False                                                            
    elif ligas['Matches'].values[x][0:4] == 'Mont' and (x % 2) == 0:            #To identify Montpellier, we check the beginning
        return 'MNT'                                                            #of the match name if Home (odd)
    elif ligas['Matches'].values[x][-5:-1: 1] == 'llie' and (x % 2) != 0:       #and the end if match visitor
        return 'MNT'
    elif ligas['Matches'].values[x][0:4] == 'Rayo' and (x % 2) == 0:            #Same with Real Valladolid (keeps RV)
        return 'RAY'                                                            #And Rayo Vallecano (changes into RAY)
    elif ligas['Matches'].values[x][-5:-1: 1] == 'ecan' and (x % 2) != 0:
        return 'RAY'
    else:
        return False
ligas['Same name and league'] = ligas['Same name and league'].apply(snal)
ligas['Teams'][ligas['Same name and league'] == 'MNT']= 'MNT'
ligas['Teams'][ligas['Same name and league'] == 'RAY']= 'RAY'
```
