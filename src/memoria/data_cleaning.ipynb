# CLEANING OUR DATASET

* In our dataset, we have to change the symbol "-", to 0's.
* Also, the column with the team names is called 'Unnamed: 2', so we will rename it as 'Teams'

```python
ligas.replace("-", 0, inplace=True)
ligas.rename(columns={'Unnamed: 2': 'Teams'}, inplace= True)
```

* We are not going to use every column. In order to make our dataset easier to work with, non-useful columns will be dropped

```python
ligas.drop(['InStat Index','Chances successful', 'Chances', 'Fouls', 'Red cards', 'Yellow cards', 'Tactics', 'Offsides', 'Corner',
            'Crosses', 'Accurate crosses, %',  'Ball recoveries in opponent\'s half', 'Ball recoveries', 'Lost balls'],
            inplace=True, axis=1)
ligas.columns
```

* Some columns values are not floats, although they might look like them, because of the % symbol. It is better to get rid of them
* After doing that, some of the values that we already converted in 0s, will become NaN's, so we will have to make them 0's again with fillna

```python
ligas['Ball possession, %'] = ligas['Ball possession, %'].str.rstrip('%').astype('float')
ligas['Chances, % of conversion'] = ligas['Chances, % of conversion'].str.rstrip('%').astype('float')
ligas['Accurate passes, %'] = ligas['Accurate passes, %'].str.rstrip('%').astype('float')
ligas['Pressing efficiency, %'] = ligas['Pressing efficiency, %'].str.rstrip('%').astype('float')

ligas.fillna(0, inplace=True)

ligas.head()
```

* To avoid duplicates in 'Matches' column, we are going to create a new column mixing every match with its date. To make it, we have to convert datetime into string (.astype(str)) and join them in the new column with ligas.insert

```python
ligas['Match date'] = ligas['Match date'].astype(str)
ligas.insert(0, 'Partidos', ligas['Matches'] + " " + ligas['Match date'])
ligas.head()
```

# VICTORY-DRAW-LOSS
In order to identify if possession heads to better results, it is necessary to know what was the result of each match for each team. With wdl function, we check the goals column to determine if the team of every row won, lost, or draw.

```python
def wdl(x):
    if (x+2) % 2 == 0:                                  #Distinction between Home and Visitor depending on its index position
        if ligas['Goals'][x] > ligas['Goals'][x+1]:     #(Home = odd, Visitor = even)
            x = 'Victory'                               #More goals than the immediate next row --> Victory
            return x                                    #Less goals than the immediate next row --> Loss
        elif ligas['Goals'][x] < ligas['Goals'][x+1]:   #Same goals --> Draw
            x = 'Loss'
            return x
        else:
            x = 'Draw'
            return x
    if (x+2) % 2 != 0:                                  #Same but with the previous row
        if ligas['Goals'][x] > ligas['Goals'][x-1]:
            x = 'Victory'
            return x
        elif ligas['Goals'][x] < ligas['Goals'][x-1]:
            x = 'Loss'
            return x
        else:
            x = 'Draw'
            return x

ligas['resultados'] = range(0, len(ligas))              #Create row with the same values as the index to iterate
ligas['resultados'] = ligas['resultados'].apply(wdl)    #Method.apply() to use it in every value of the column
ligas.head(4)
```

## Duplicate teams: different league
It was found that some teams shared the same short name, so we had to separate their names again.
If teams play in different leagues we can change the names by conditioning the name to the league they play in

```python
ligas['Teams'][(ligas['Teams'] == 'MON') & (ligas['Competition'] == 'Serie A')]='MNZ'
ligas['Teams'][(ligas['Teams'] == 'BRE') & (ligas['Competition'] == 'Premier League')]='BRF'
ligas['Teams'][(ligas['Teams'] == 'BRE') & (ligas['Competition'] == 'Serie A')]='BRS'
```

## Duplicate teams: same league

When teams with the same name play in the same league, we will check the match name to make them different again. To get that, it was necessary to create a column: 'Same name and league', and a function 'snal'.

We have two cases of that, Monaco-Montpellier (both MON), and Real Valladolid-Rayo Vallecano (both RV)

```python
ligas['Same name and league'] = range(0, len(ligas))        #We needed to iterate over a column with the same values as the index

def snal(x):
    if ligas['Teams'][x] != 'MON' and ligas['Teams'][x] != 'RV':                #Monaco keeps MON so we will call Montpellier MNT
        return False                                                            
    elif ligas['Matches'].values[x][0:4] == 'Mont' and (x % 2) == 0:            #To identify Montpellier, we check the beginning
        return 'MNT'                                                            #of the match name if Home (odd)
    elif ligas['Matches'].values[x][-5:-1: 1] == 'llie' and (x % 2) != 0:       #and the end if match visitor
        return 'MNT'
    elif ligas['Matches'].values[x][0:4] == 'Rayo' and (x % 2) == 0:            #Same with Real Valladolid (keeps RV)
        return 'RAY'                                                            #And Rayo Vallecano (changes into RAY)
    elif ligas['Matches'].values[x][-5:-1: 1] == 'ecan' and (x % 2) != 0:
        return 'RAY'
    else:
        return False
ligas['Same name and league'] = ligas['Same name and league'].apply(snal)
ligas['Teams'][ligas['Same name and league'] == 'MNT']= 'MNT'
ligas['Teams'][ligas['Same name and league'] == 'RAY']= 'RAY'
```

* Total possession should be 8006 matches * 100 = 800800, we should investigate where is that 70 points difference

```python
ligas['posesioncomp'] = range(0, len(ligas))
def hola(x):
    if (x+2) % 2 == 0:       
        return ligas['Ball possession, %'][x] + ligas['Ball possession, %'][x+1]                               
    elif (x+2) % 1 == 0:
        return ligas['Ball possession, %'][x] + ligas['Ball possession, %'][x-1]
    else:
        return False
    
ligas['posesioncomp'] = ligas['posesioncomp'].apply(hola)
ligas[ligas['posesioncomp'] != 100]
```

We can see, in 70 matches along the years, percentages sumed up 101 per cent. Not a critical damage for our data
