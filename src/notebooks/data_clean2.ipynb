```python
ucl.replace("-", 0, inplace=True)
ucl.rename(columns={'Unnamed: 2': 'Teams'}, inplace= True)
ucl.head()
```

```python
ucl.drop(['InStat Index','Chances successful', 'Chances', 'Fouls', 'Red cards', 'Yellow cards', 'Tactics', 'Offsides', 'Corner',
            'Crosses', 'Accurate crosses, %',  'Ball recoveries in opponent\'s half', 'Ball recoveries', 'Lost balls'],
            inplace=True, axis=1)
ucl.columns
```

```python
ucl['Ball possession, %'] = ucl['Ball possession, %'].str.rstrip('%').astype('float')
ucl['Chances, % of conversion'] = ucl['Chances, % of conversion'].str.rstrip('%').astype('float')
ucl['Accurate passes, %'] = ucl['Accurate passes, %'].str.rstrip('%').astype('float')
ucl['Pressing efficiency, %'] = ucl['Pressing efficiency, %'].str.rstrip('%').astype('float')

ucl.fillna(0, inplace=True)

ucl.head()
```

```python
ucl['Match date'] = ucl['Match date'].astype(str)
ucl.insert(0, 'Partidos', ucl['Matches'] + " " + ucl['Match date'])
ucl.head()
```

```python
def wdl(x):
    if (x+2) % 2 == 0:                                  #Distinction between Home and Visitor depending on its index position
        if ucl['Goals'][x] > ucl['Goals'][x+1]:     #(Home = odd, Visitor = even)
            x = 'Victory'                               #More goals than the immediate next row --> Victory
            return x                                    #Less goals than the immediate next row --> Loss
        elif ucl['Goals'][x] < ucl['Goals'][x+1]:   #Same goals --> Draw
            x = 'Loss'
            return x
        else:
            x = 'Draw'
            return x
    if (x+2) % 2 != 0:                                  #Same but with the previous row
        if ucl['Goals'][x] > ucl['Goals'][x-1]:
            x = 'Victory'
            return x
        elif ucl['Goals'][x] < ucl['Goals'][x-1]:
            x = 'Loss'
            return x
        else:
            x = 'Draw'
            return x

ucl['resultados'] = range(0, len(ucl))              #Create row with the same values as the index to iterate
ucl['resultados'] = ucl['resultados'].apply(wdl)    #Method.apply() to use it in every value of the column
ucl.head(4)
```

```python
ucl.groupby(['resultados'])[['Ball possession, %']].mean()
ucl['Ball possession, %'].sum()
```

```python
ucl['posesioncomp'] = range(0, len(ucl))
def pos_ucl(x):
    if (x+2) % 2 == 0:       
        return ucl['Ball possession, %'][x] + ucl['Ball possession, %'][x+1]                               
    elif (x+2) % 1 == 0:
        return ucl['Ball possession, %'][x] + ucl['Ball possession, %'][x-1]
    else:
        return False
    
ucl['posesioncomp'] = ucl['posesioncomp'].apply(pos_ucl)
ucl[ucl['posesioncomp'] != 100]
```
