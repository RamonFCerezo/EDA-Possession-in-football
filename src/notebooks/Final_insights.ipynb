```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
ligas = pd.read_csv("data/LIGAS.csv", sep= '|')
```

```python
elim = pd.read_csv("data/ELIMINATORIAS.csv", sep = '|')
```

```python
melted = pd.read_csv("data/ALL.csv", sep='|')
```

```python
gyr = ['#F7DE3A', '#FF0000', '#7CFC00']
```

```python
data = ligas[(ligas['Ball possession, %'] > 50)].groupby(['resultados'])['Ball possession, %'].count()

plt.figure(figsize=(8,8))
plt.title('Resultados - Posesión > 50% (5 grandes ligas)', fontsize= 20)
plt.pie(data.values,
        labels=data.index,
        autopct='%1.2f%%',
        colors= gyr,
        explode = [0.005, 0.005, 0.005],
        textprops={'fontsize': 18});

p=plt.gcf();
```
